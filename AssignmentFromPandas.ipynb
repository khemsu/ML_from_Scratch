# Data types

```python
df_null_types = df.isnull().any().dtypes
print(df_null_types)
```

```python
#changing year of release to int
x = df['Year_of_Release'].mean()
df['Year_of_Release'].fillna(x, inplace=True)
df['Year_of_Release'] = df['Year_of_Release'].astype(int)
df.dtypes
```

```python
#identifying null values
isnull  = df.isnull()
print(isnull)
```

```python
#counting the null  by column
isnull_count= isnull.sum()
print("Total number of null are:\n",isnull_count)
```

```python
#total null values in datasets
# isnull vaneko null values ho ani tesko null le column wise null value dinxa ani tesko sum le total null value dinxa
isnull_sum = isnull.sum().sum()
print("Total number of null values are:\n",isnull_sum)
```

Fill Nan values in coulmns with Unknown

```python
columns_to_fill = ["Name", "Year_of_Release", "Genre", "Publisher",
                   "Critic_Score", "Critic_Count", "User_Score",
                   "User_Count", "Developer", "Rating"]

df[columns_to_fill] = df[columns_to_fill].fillna("Unknown")
print(df)
```

```python
df_modified = df.to_csv("modified_dataset.csv", index=False)
print(df_modified)
```
