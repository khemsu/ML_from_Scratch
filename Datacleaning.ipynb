```python
isnull  = df.isnull()
print(isnull)
```

```python
isnull_count= isnull.sum()
print("Total number of null are:", isnull_count)
```

```python
total_null = df.isnull().sum().sum()
print("Total number of null values are:", total_null)
```

```python
#replace single qoute from dataframe
df['Date'] = df['Date'].replace("'", "", regex=True)
print(df)
```

```python
#handle null values, return dataframe with no empty cells
#we can use df.dropna(), it removes empty cells but will not change the dataframes
#if we want dataframe to be changed, simply use inplace=True

new_df = df.dropna()
print(new_df)
```

```python
isnull_count_newdf = new_df.isnull().sum()
print(isnull_count_newdf)
```

```python
#another way to handle null value that can directly handle null values in dataframe without creating new_df
#130 is replaced in null
df.fillna(130, inplace=True)
df
```

```python
#another way of handling null
df = pd.read_csv(r"/content/pandas_dataset.csv")
df['Calories'].fillna(130, inplace=True)
df
```

```python
#fill specific with mean, median, and mode
df = pd.read_csv(r"/content/pandas_dataset.csv")
x = df['Calories'].mean()
df['Calories'].fillna(x, inplace=True)
df
```
