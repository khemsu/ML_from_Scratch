Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## line graph

```python
gas = pd.read_csv('matplotlib/gas_prices.csv')

plt.figure(figsize=(8,5), dpi=200)

plt.title('Gas price over Time')

#here Year and USA are the data of the csv file
# also we can add other attributes of line in plt.plot
# plt.plot(gas.Year, gas.USA, 'r.-',label='USA', )
# plt.plot(gas.Year, gas.Canada, 'b.-' , label='Canada')
# plt.plot(gas.Year, gas['South Korea'],'g.-', label='South Korea')

#for all countries
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label='country')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollar')
        

plt.legend()

plt.show()
```

## pie chart

```python
fifa = pd.read_csv('matplotlib/fifa_data.csv')
```

```python
fifa.head(5)
```

```python
#.shape is use to access the row that has preferred foot left , right & [0] represents rows, if shape[1] refresents coloumn
left= fifa[fifa['Preferred Foot'] == 'Left'].shape[0]
Right= fifa[fifa['Preferred Foot'] == 'Right'].shape[0]

labels= ['left', 'right']

plt.pie([left,Right], labels = labels, autopct='%.2f %%')

plt.title('Foot preferences of Fifa players')
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa[fifa.Weight < 125].shape[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].shape[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].shape[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].shape[0]
heavy = fifa[fifa.Weight >= 200].shape[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

# pctdiatance is distance of percentage from the center of pie chart

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
