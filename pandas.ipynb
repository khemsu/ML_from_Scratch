Loading data in pandas

```python
import pandas as pd

df = pd.read_csv('/content/pokemon_data.csv')
print(df.head(3))
```

Reading row or specific rows

```python
print(df.loc[0])
```

```python
#Reading row 0 and 5
print(df.loc[[0,5]])
```

```python
#to read each column
print(df[['Name', 'Type 1', 'HP']])
```

```python
#to read each row
print(df.iloc[3:5])
```

TO read specific location

```python
print(df.iloc[3,1])
```

```python
#to find specific rows based on specific textual information or interger
# df.loc[df['Type 1'] == "Fire"]
df.loc[(df['Type 1'] == "Fire") & (df['Type 2'] == "Flying")]
```

# **Soritng/Describing Data**


```python
df.describe()
```

# **Sort the data by Name**

```python
df.sort_values('Name')
```

```python
#Ascending by the type 1 and HP
df.sort_values(['Type 1', 'HP'], ascending = True) # Change "True" to True
```

```python
#to ascend by type and descend by the HP
df.sort_values(['Type 1', 'HP'], ascending = [1,0])
```

Making changes to the data

```python
#adding "Total" in the data set
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
#Dropping the column
df.drop(columns = ['Total'])
```

```python
# adding the column   [all rows, columns]
df['Total'] = df.iloc[      :  , 4:10].sum(axis=1)
df.head(5)
```

Saving Data

```python
df.to_csv('modified csv', index = False)
```

# Using regex
Regex is useful for searching patterns in data sets , more flexible, and helps in searchings more complex patterns.

```python
import re
df.loc[df['Type 1'].str.contains('fire|grass' , flags=re.I , regex =True)]
```
